```python
import pandas as pd
```

```python
df = pd.read_csv('SaaS-Sales.csv')
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.nunique()
```

```python
df.isna().sum()
```

```python
num_cols = df.select_dtypes(include=['float','int']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df[['Sales','Profit','Discount']].describe()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df['Year'] = df['Order Date'].dt.year
```

```python
df.head(5)
```

```python
df['Quarter'] = df['Order Date'].dt.to_period('Q')
```

```python
df.head(2)
```

```python
df['Month'] = df['Order Date'].dt.month
```

```python
import matplotlib.pyplot as plt

sales_time = df.groupby('order_date')['sales'].sum()
sales_time.plot(title="Sales Over Time")
top_sales = df.groupby('product')['sales'].sum().sort_values(ascending=False).head(10)
plt.show()
```

```python
profit_time = df.groupby('order_date')['profit'].sum()
profit_time.plot(title="Profit Over Time")
top_profit = df.groupby('product')['profit'].sum().sort_values(ascending=False).head(10)
plt.show()
```

```python
top_profit
```

```python
top_sales.plot(kind='barh', title="Top 10 Products by Sales")
plt.show()
```

```python
top_profit.plot(kind='bar', title='Top profit by items')
```

```python
df.columns = (df.columns
    .str.strip()
    .str.lower()
    .str.replace(' ','_')
)
```
